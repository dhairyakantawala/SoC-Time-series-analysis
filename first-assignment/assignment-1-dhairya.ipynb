```python
#1
#loading the dataset

import pandas as pd
df = pd.read_csv('Global_Superstore2.csv', encoding='latin1')
```

```python
#display
print(df.head())
```

```python
# Summary of the dataset
print(df.info())
```

```python
# Basic statistical details
print(df.describe())
```

```python
# Count missing values in each column
missing_values = df.isnull().sum()

# Display columns with missing values
print("Columns with missing values:")
print(missing_values[missing_values > 0])
```

```python
# Drop the "Postal Code" column
df.drop(columns=['Postal Code'], inplace=True)
```

```python
# Remove duplicate rows
df.drop_duplicates(inplace=True)
```

```python
print(df.info())
```

```python
# Convert "Order Date" column to datetime type
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Convert "Ship Date" column to datetime type
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
print(df.head())
```

```python
# Descriptive statistics for numerical columns
numerical_stats = df.describe()
print(numerical_stats)
```

```python
# List of categorical column names to analyze
categorical_columns = ['Ship Mode', 'Segment', 'City', 'Market', 'Region', 'Category', 'Sub-Category', 'Order Priority']

# Generate summary statistics for each categorical column
for column_name in categorical_columns:
    print(f"Summary statistics for '{column_name}':")
    print(df[column_name].value_counts())
    print()
```

```python
import matplotlib.pyplot as plt

# Histogram for Sales
plt.figure(figsize=(10, 6))
plt.hist(df['Sales'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of Sales')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()


# Box plot for Sales
plt.figure(figsize=(8, 6))
plt.boxplot(df['Sales'])
plt.title('Box Plot of Sales')
plt.ylabel('Sales')
plt.show()
```

```python
# Histogram for Quantity
plt.figure(figsize=(10, 6))
plt.hist(df['Quantity'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of Quantity')
plt.xlabel('Quantity')
plt.ylabel('Frequency')
plt.show()


# Box plot for Quantity
plt.figure(figsize=(8, 6))
plt.boxplot(df['Quantity'])
plt.title('Box Plot of Quantity')
plt.ylabel('Quantity')
plt.show()
```

```python
# Histogram for Discount
plt.figure(figsize=(10, 6))
plt.hist(df['Discount'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of Discount')
plt.xlabel('Discount')
plt.ylabel('Frequency')
plt.show()


# Box plot for Discount
plt.figure(figsize=(8, 6))
plt.boxplot(df['Discount'])
plt.title('Box Plot of Discount')
plt.ylabel('Discount')
plt.show()
```

```python
import seaborn as sns

# Count plot for Category
plt.figure(figsize=(10, 6))
sns.countplot(x='Category', data=df, palette='pastel')
plt.title('Count Plot of Category')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()
```

```python
# Count plot for Sub-Category
plt.figure(figsize=(10, 6))
sns.countplot(x='Sub-Category', data=df, palette='pastel')
plt.title('Count Plot of Sub-Category')
plt.xlabel('Sub-Category')
plt.ylabel('Count')
plt.show()
```

```python
# Scatter plot of Sales vs. Profit
plt.figure(figsize=(8, 6))
plt.scatter(df['Sales'], df['Profit'], color='skyblue', alpha=0.5)
plt.title('Scatter Plot of Sales vs. Profit')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.show()
```

```python
# Correlation matrix
# # Select only numerical columns for correlation matrix
numerical_columns = df.select_dtypes(include=['float64', 'int64'])

# Correlation matrix
correlation_matrix = numerical_columns.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
# Pair plot
sns.pairplot(df[['Sales', 'Profit', 'Quantity', 'Discount']])
plt.title('Pair Plot of Sales, Profit, Quantity, and Discount')
plt.show()
```

```python
# Bar plot of Sales by Category
plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Sales', data=df, palette='pastel')
plt.title('Bar Plot of Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.show()
```

```python
# Box plot of Profit by Region
plt.figure(figsize=(10, 6))
sns.boxplot(x='Region', y='Profit', data=df, palette='pastel')
plt.title('Box Plot of Profit by Region')
plt.xlabel('Region')
plt.ylabel('Profit')
plt.show()
```

```python
# Pivot table for average Sales and Profit by Category and Sub-Category
pivot_table = df.pivot_table(index=['Category', 'Sub-Category'], values=['Sales', 'Profit'], aggfunc='mean')

# Display pivot table
print(pivot_table)
```

```python
# Group-by operation for average Sales and Profit by Category and Sub-Category
grouped_data = df.groupby(['Category', 'Sub-Category']).agg({'Sales': 'mean', 'Profit': 'mean'})

# Display grouped data
print(grouped_data)
```

```python
# Calculate Profit Margin
df['Profit Margin'] = (df['Profit'] / df['Sales']) * 100

# Calculate Sales per Customer
sales_per_customer = df.groupby('Customer ID')['Sales'].sum()
df['Sales per Customer'] = df['Customer ID'].map(sales_per_customer)

# Calculate Order Processing Time
df['Order Processing Time'] = (df['Ship Date'] - df['Order Date']).dt.days
```
