```python
import os
import csv
from PIL import Image
import numpy as np

# Define the image directory and the output CSV file
image_dir = 'pizza'
output_csv = 'pizza.csv'
image_size = (28, 28)  # Resize images to 28x28 pixels

# Function to process images and write to CSV
def images_to_csv(image_dir, output_csv, image_size):
    # Create the header for the CSV file
    header = []
    for i in range(image_size[0] * image_size[1]):
        header.extend([f'R{i+1}', f'G{i+1}', f'B{i+1}'])
    
    # Open the CSV file for writing
    with open(output_csv, mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(header)  # Write the header

        # Iterate through each image file in the directory
        for filename in os.listdir(image_dir):
            if filename.endswith('.png') or filename.endswith('.jpg') or filename.endswith('.jpeg'):
                # Read and resize the image
                img_path = os.path.join(image_dir, filename)
                img = Image.open(img_path).resize(image_size)
                
                # Convert image to numpy array and separate color channels
                img_array = np.array(img)
                img_flattened = img_array.reshape(-1, 3)  # Shape (n_pixels, 3)
                img_values = img_flattened.flatten()  # Shape (n_pixels*3,)

                # Write the image pixel values to the CSV
                writer.writerow(img_values)

# Run the function
images_to_csv(image_dir, output_csv, image_size)
```
