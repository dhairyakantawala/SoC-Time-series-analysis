```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
pizza_df = pd.read_csv('pizza.csv')
not_pizza_df = pd.read_csv('not_pizza.csv')

pizza_df['is_pizza'] = True
not_pizza_df['is_pizza'] = False

pizza_df = pizza_df[['is_pizza'] + [col for col in pizza_df.columns if col != 'is_pizza']]
not_pizza_df = not_pizza_df[['is_pizza'] + [col for col in not_pizza_df.columns if col != 'is_pizza']]
```

```python
pizza_df
```

```python
not_pizza_df
```

```python
df = pd.concat([pizza_df, not_pizza_df], ignore_index=True)
```

```python
df
```

```python
X = df.drop(columns='is_pizza', axis=1)
Y = df['is_pizza']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
X_test_prediction_LR = model.predict(X_test)
```

```python
acc = accuracy_score(Y_test, X_test_prediction_LR)
print(f"accuracy of the Logistic Regression is: {acc}")
```

```python
from sklearn import svm
```

```python
model_RBF = svm.SVC(degree=3, gamma="scale", kernel="rbf")
model_RBF.fit(X_test, Y_test)
```

```python
X_test_prediction_SVC = model_RBF.predict(X_test)
```

```python
acc = accuracy_score(Y_test, X_test_prediction_SVC)
print(f"accuracy of the SVC model is: {acc}")
```
